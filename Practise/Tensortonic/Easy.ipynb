### 1. Implement Sigmoid in NumPy

```python
import numpy as np

def sigmoid(x):
    """
    Vectorized sigmoid function.
    """
    x = np.asarray(x, dtype=float)
    return 1 / (1 + np.exp(-x))
sigmoid(np.array([-1, 0, 1]))
```

## 2. Matrix Transpose


```python
import numpy as np

def matrix_transpose(A):
    """
    Return the transpose of matrix A (swap rows and columns).
    """
    A = np.asarray(A, dtype=float)
    rows, cols = A.shape
    transposed = np.empty((cols, rows), dtype=A.dtype)

    for i in range(rows):
        for j in range(cols):
            transposed[j, i] = A[i, j]

    return transposed
A = np.array([[1, 2], [3, 4]])
print(matrix_transpose(A))
```

## 3.Implement Gradient Descent for a 1D Quadratic


```python
def gradient_descent_quadratic(a, b, c, x0, lr, steps):
    """
    Return final x after 'steps' iterations.
    """
    x = float(x0)
    for _ in range(steps):
        grad = 2 * a * x + b
        x = x - lr * grad
    return float(x)

gradient_descent_quadratic(a=1, b=-4, c=4, x0=0, lr=0.1, steps=100)
```
