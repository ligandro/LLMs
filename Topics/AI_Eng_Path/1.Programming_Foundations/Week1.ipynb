### Practice 2: Dictionaries and Lists
Goal: transform raw data into a clean summary structure.

```python
students = [
    {"name": "Ari", "scores": [85, 90, 88]},
    {"name": "Sam", "scores": [78, 80, 74]},
    {"name": "Lee", "scores": [92, 95, 91]},
]

summary = {}
for student in students:
    avg_score = sum(student["scores"]) / len(student["scores"])
    summary[student["name"]] = round(avg_score, 2)

print("Average scores by student:")
print(summary)

high_performers = [name for name, avg in summary.items() if avg >= 85]
print("High performers:", high_performers)
```

### Practice 3: Function Design
Goal: write small functions that each do one job clearly.

```python
def normalize_text(text: str) -> str:
    return " ".join(text.strip().lower().split())


def word_count(text: str) -> int:
    if not text:
        return 0
    return len(text.split())


def summarize_text(text: str) -> dict:
    cleaned = normalize_text(text)
    return {
        "cleaned": cleaned,
        "word_count": word_count(cleaned),
    }

sample = "  Python functions should do one thing well.  "
print(summarize_text(sample))
```
